## Using map backgrounds in matplotlib

In this simple example we will show how to make map backgrounds using the "mpl_toolkit".  This module provides a set of routines called "Basemap", which we will use to draw coastlines, shade land/ocean, etc., much like we did with GMT earlier.  Here we just plot the "daylight" and "night" as different shading on the globe.

```python
# import all our modules; the new one here is the Basemap routines from mpl_toolkit
import numpy as np
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Next, we define a map instance (m) defined from basemap.  Basemap (again like GMT) allows
#  for a ton of different projections, and each one takes different arguments.

# E.g., Miller projection just needs the center longitude, and we'll use the Pacific as 
#  our map center
m = Basemap(projection='mill',lon_0=180)

# We can now proceed to plot things like coastlines
m.drawcoastlines()
#  and lat/lon lines
m.drawparallels(np.arange(-90,90,30),labels=[1,0,0,0])
m.drawmeridians(np.arange(map.lonmin,map.lonmax+30,60),labels=[0,0,0,1])

#  and fill land ('coral' colored) and ocean ('aqua')
m.drawmapboundary(fill_color='aqua')
m.fillcontinents(color='coral',lake_color='aqua')

# Finally, just to see a cool feature of datetime, we can get the current time
date = datetime.utcnow()

# and then shade night
CS = m.nightshade(date)

plt.title('Day/Night Map for %s (UTC)' % date.strftime("%d %b %Y %H:%M:%S"))
```
